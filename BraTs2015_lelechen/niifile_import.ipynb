```python
#niiファイルの読み込み

nii_path = 'VSD.Brain.XX.O.MR_Flair.35533.nii'
mha_path = '/Users/ryoheieguchi/Desktop/Class/BRATS2015_Training/HGG/brats_tcia_pat105_0001/VSD.Brain.XX.O.MR_Flair.35533/VSD.Brain.XX.O.MR_Flair.35533.mha'

img = sitk.ReadImage(mha_path)
sitk.WriteImage(img, nii_path)
```

```python
nii0=nib.load(nii_path)
img0=nii0.get_data()
print(img0.shape) # 三次元配列の大きさ表示
plt.imshow(img0[:,:,img0.shape[2]/2].T,cmap='gray',origin='lower')
#plt.imshow(img0[1,1,1],cmap='gray',origin='lower')
plt.show()
```

```python
img = nib.load(nii_path)
img_data = img.get_fdata()
print(img_data.shape)

gs = gridspec.GridSpec(2, 4, width_ratios=[1.17, 1, 1.17])
plt.figure(figsize=(8,4))

for i in range(7):
    plt.subplot(gs[i])
    plt.imshow(img_data[:, :, i*20], cmap="gray")
plt.tight_layout()
plt.show()
```
